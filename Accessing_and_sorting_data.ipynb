```python
#Sorting Data
parquet.sort_values(['year','athlete_id'],ascending=False)
```

```python
for index,row in parquet.iterrows():
  print(index)
  print(row)
  if index>5:
    break
```

```python
#filtering Data Using Conditions
parquet.loc[parquet['tied']==False]
parquet.loc[(parquet['place']==2) & (parquet['year'].between(1981, 1984)) | (parquet['type']=='Summer')]
parquet.loc[(parquet['event'].str.contains("Men")) & (parquet['event'].str.contains("Singles")) & (parquet['year'].between(1981, 1990)) & (parquet['place']==2) & (parquet['type']=='Summer') ]
parquet.loc[(parquet['place'].isin([4,5,6])) & (parquet['year'].between(1998,2026)) & (parquet['type']=='Summer') & (parquet['discipline']=='Athletics') & (parquet['event'].str.contains("Men"))]
```

```python
#Adding and Removing Columns
parquet['Details']=parquet['type']+" - "+parquet['discipline']
parquet.sample(20)
```

```python
parquet.drop(columns=['Details'],inplace=True)
```
