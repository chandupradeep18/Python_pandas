```python
#Loading Data From parquet File Into DataFrame
parquet=pd.read_parquet('/content/drive/MyDrive/Data Sources/results.parquet')
print(parquet.columns)
```

```python
parquet.head(10) #Show first 10 row of The Data
```

```python
parquet.tail(10) #Show last 10 row of the data
```

```python
parquet.sample(20) #Show Ramdom 20 Row from The data
```

```python
#loc Locate Data Using Index Value or Values and Select Required Colomn
parquet.loc[9018] #get data using Index Value and You can Select Required Colomn Name or Leave it for all Colomn
```

```python
parquet.loc[9018,['year','type','athlete_id']]
```

```python
parquet.loc[3618:5499:10,['year','type','athlete_id']]
```

```python
#iloc Locate Data Using Index Value or Values and Select Required Colomn using Value not Name
parquet.iloc[9018,3]
```

```python
parquet.iloc[9018,[1,2,3,4,5]]
```

```python
parquet.iloc[1629:5499,[1,2,3,4,5]]
```

```python
#using loc we Can Modify data In Data Set
print(parquet.loc[9018])
parquet.loc[9018,'tied']=True
parquet.loc[9018,'medal']='Gold'
print(parquet.loc[9018])
#using iloc we Can Modify data In Data Set
print(parquet.loc[9018])
parquet.loc[9018,9]=False
parquet.loc[9018,10]=None
print(parquet.loc[9018])
```

```python
parquet.at[9018,'year'] #get Specific Value Using Coloumn Name
```

```python
parquet.iat[9018,1] #get Specific Value Using Coloumn Value
```
