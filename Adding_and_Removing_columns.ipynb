```python
import pandas as pd
import numpy as np
```

```python
current_date = pd.Timestamp.now()
amazon_sales=pd.read_csv('/content/drive/MyDrive/Data Sources/amazon_sales.csv')
amazon_sales.sample(10)
bios=pd.read_csv('/content/drive/MyDrive/Data Sources/bios.csv')
bios.sample(10)
olympics=pd.read_parquet('/content/drive/MyDrive/Data Sources/results.parquet')
olympics.sample(10)
```

```python
#Adding New Column to Data Frame
bios['first_name']=bios['name'].str.split(' ').str[0]
bios['last_name']=bios['name'].str.split(' ').str[1]
bios['status']=np.where(bios['died_date'].isna(),'Alive','Died')
bios['Note']=np.where(bios['height_cm'].isna(),'Height is Required','')
bios.sample(5)
```

```python
#Removing a Columns from DataFrame
bios.drop(columns=['Note'],inplace=True)
bios.sample(10)
```

```python
bios['born_date']=pd.to_datetime(bios['born_date'])
bios['died_date']=pd.to_datetime(bios['died_date'])
```

```python
bios['age']=np.where(bios['died_date'].isna(),
                     current_date.year-bios['born_date'].dt.year,
                     current_date.year-bios['died_date'].dt.year)
```

```python
bios[(bios['age']>80) & (bios['status']!='Alive')]
```

```python
bios['height_category']=bios['height_cm'].apply(lambda x : 'short' if x<165 else('Average' if x<185 else 'Tall'))
bios.sample(20)
```

```python
def weight(row):
  if row['weight_kg']>80:
    return 'Over Weight'
  else:
    return 'Normal Weight'
bios['weight_category']=bios.apply(weight,axis=1)
bios.sample(18)
```

```python
#save to Data Frame to CSV File
bios.to_csv('/content/drive/MyDrive/Data Sources/olympicsbios.csv',index=False)
```

```python
bios[(bios['age']>80) & (bios['status']=='Alive')].to_csv('/content/drive/MyDrive/Data Sources/olympicsbios_alive.csv',index=False)
```

```python
bios[(bios['age']>80) & (bios['status']!='Alive')].to_csv('/content/drive/MyDrive/Data Sources/olympicsbios_died.csv',index=False)
```

```python
bios[(bios['born_country']=='IND')].to_csv('/content/drive/MyDrive/Data Sources/olympicsbios_india.csv',index=False)
```
